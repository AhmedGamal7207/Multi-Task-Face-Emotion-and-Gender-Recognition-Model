Dataset Creation

```python
import numpy as np
import cv2
import os
```

```python
dataset_path = 'Images'
face_path = 'Images_Faces'
persons = ['maya','mona'] 

face_detect = cv2.CascadeClassifier('face.xml') # Pretrained model to detect faces
```

```python
for person in persons:
    person_path = os.path.join(dataset_path, person) # the folder path of the person images' folder
    face_path = os.path.join(face_path,person) # the folder path of the person faces' (output) folder 
    # Iterate over each image in the person's folder
    for filename in os.listdir(person_path): 
        if filename.endswith('.jpg'): # ensuring it's an image
            # Load image
            image_path = os.path.join(person_path, filename) # the full path of that image
            image = cv2.imread(image_path) # loading the image
            face = face_detect.detectMultiScale(image,1.3,5) # detecting the face in the image
            old_filename = filename # saving the old file name (maya-happy-1-301.jpg)
            name,emotion,finger,count = filename.split("-") # splitting to get the labels
            gender = "female" # setting the gender to female
            filename = f"{name}-{emotion}-{gender}-{count}" # the new file name
            for x,y,w,h in face:
                cv2.imwrite(os.path.join(face_path,filename),image[y:y+h,x:x+w]) # saving only the face image
```

To have balanced data:

```python
import os
import random

def randomly_delete_images(folder_path, target_count):
    # Get the list of all images in the folder
    all_images = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]

    # Calculate the number of images to delete
    images_to_delete = len(all_images) - target_count

    # If the current count is equal to or less than the target count, do nothing
    if images_to_delete <= 0:
        print(f"The folder already contains {target_count} images. No images will be deleted.")
        return

    # Randomly select images to delete
    images_to_delete_list = random.sample(all_images, images_to_delete)

    # Delete the selected images
    for image in images_to_delete_list:
        image_path = os.path.join(folder_path, image)
        os.remove(image_path)
        print(f"Deleted: {image}")

    print(f"Total images in the folder: {len(os.listdir(folder_path))}")

# Replace 'your_folder_path' with the actual path of your folder
folder_path = 'Images_Faces/mona/'

# Set the target count of images
target_count = 1204

# Call the function to randomly delete images
randomly_delete_images(folder_path, target_count)
```
